```python
import scipy.optimize
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_values = []

def f(x):
    global x_values
    x_values.append(x)
    return (4.0 + x)**2 + np.random.random()
```

```python
scipy.optimize.minimize(f, x0=0, method="BFGS")
```

```python
plt.plot(x_values)
print("number of function evaluation", len(x_values))
```

```python
import torch
```

```python
n_epoch = 500

x = torch.nn.parameter.Parameter(torch.zeros(1))
opt = torch.optim.Adam([x, ], lr=0.5)

x_vals = []

for epoch in range(n_epoch):
    
    loss = f(x)
    opt.zero_grad()
    loss.backward()
    
    with torch.no_grad():
        x_vals.append(x.item())
    
    opt.step()
```

```python
print("final value", x.item())
```

```python
plt.plot(x_vals)
```

```python
import optuna
```

```python
x_vals = []

def objective(trial):
    x = trial.suggest_float('x', -10, 10)
    global x_vals
    x_vals.append(x)
    return f(x)

study = optuna.create_study()
study.optimize(objective, n_trials=200)
```

```python
from stable_baselines3 import PPO
import gym
from gym import spaces
from stable_baselines3.common.env_checker import check_env
```

```python
class CustomEnv(gym.Env):
    """Custom Environment that follows gym interface"""
    metadata = {'render.modes': ['human']}

    def __init__(self):
        super(CustomEnv, self).__init__()
        self.action_space = spaces.Box(low=-1, high=1, shape=(1, ))
        self.observation_space = spaces.Discrete(1)
        # self.observation_space = spaces.Box(low=-1.0, high=1.0, shape=(1, )) # spaces.Box(low=obs_min, high=obs_max, shape=(1, ))
        # self.obs_ref = self.observation_space.sample()
    
    def step(self, action):
        x = action[0] * 10.0
        # observation = np.zeros_like(self.obs_ref)
        observation = 0
        reward = - f(x)
        done = False
        info = {}
        return observation, reward, done, info

    def reset(self):
        # observation = np.zeros_like(self.obs_ref)
        observation = 0
        return observation  # reward, done, info can't be included

    def render(self, mode='human'):
        pass

    def close(self):
        pass
```

```python
check_env(CustomEnv())
```

```python
env = CustomEnv()

model = PPO('MlpPolicy', env, verbose=1)
model.learn(total_timesteps=30000, log_interval=np.inf)

obs = env.reset()

action, _state = model.predict(obs, deterministic=True)
print(action*10)

# for i in range(1000):
#     action, _state = model.predict(obs, deterministic=True)
#     obs, reward, done, info = env.step(action)
#     # env.render()
#     if done:
#         obs = env.reset()
```
